
# DETECTOR FUSION
# YOLOv8 + Faster R-CNN + RetinaNet (Detectron2)


```python
# ============================================================
# IMPORTS
# ============================================================

import torch
import cv2
import numpy as np

from ultralytics import YOLO

from torchvision.models.detection import (
    fasterrcnn_resnet50_fpn_v2
)

from torchvision.models.detection.faster_rcnn import (
    FastRCNNPredictor
)

from torchvision.ops import nms

from torchvision import transforms

from ensemble_boxes import weighted_boxes_fusion

# Detectron2
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2 import model_zoo
```

## Upload Image

```python
# ============================================================
# UPLOAD IMAGE
# ============================================================

from google.colab import files

uploaded = files.upload()

IMAGE_PATH = next(iter(uploaded))

print(f"Uploaded image: {IMAGE_PATH}")

# ============================================================
# LOAD IMAGE
# ============================================================

image_bgr = cv2.imread(IMAGE_PATH)

if image_bgr is None:
    raise ValueError(f"Could not read image: {IMAGE_PATH}")

image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

H, W = image_rgb.shape[:2]
```

## Paths

```python
# ============================================================
# CONFIG
# ============================================================

YOLO_WEIGHTS = "/content/YOLO.pt"

FASTER_WEIGHTS = "/content/FASTER_R_CNN.pth"

RETINA_WEIGHTS = "/content/RetinaNet.pth"


RESNET50_WEIGHTS = "/content/RESNET50.pth"

ConvNext_WEIGHTS = "/content/ConvNext.pth"

Efficient_WEIGHTS = "/content/Efficient.pth"


DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
```

## Thresholds

```python
# ============================================================
# THRESHOLDS
# ============================================================

YOLO_CONF = 0.153

FASTER_CONF = 0.65

RETINA_CONF = 0.10

FINAL_CONF = 0.15

WBF_IOU = 0.90

MIN_BOX_AREA = 1000
```

## Models

```python


# ============================================================
# CLASS NAMES
# ============================================================

CLASS_NAMES = {
    0: "Cardboard",
    1: "Clothes",
    2: "Garbage_bag",
    3: "Glass",
    4: "Metal",
    5: "Organic",
    6: "Plastics"
}

NUM_CLASSES = len(CLASS_NAMES) + 1

# ============================================================
# LOAD YOLO
# ============================================================

print("Loading YOLO...")

yolo_model = YOLO(YOLO_WEIGHTS)

print("YOLO loaded successfully.")

# ============================================================
# LOAD FASTER R-CNN
# ============================================================

print("Loading Faster R-CNN...")

faster_model = fasterrcnn_resnet50_fpn_v2(
    weights=None,
    weights_backbone=None
)

in_features = faster_model.roi_heads.box_predictor.cls_score.in_features

faster_model.roi_heads.box_predictor = FastRCNNPredictor(
    in_features,
    NUM_CLASSES
)

checkpoint = torch.load(
    FASTER_WEIGHTS,
    map_location=DEVICE
)

faster_model.load_state_dict(
    checkpoint["model_state_dict"]
)

faster_model.to(DEVICE)
faster_model.eval()

print("Faster R-CNN loaded successfully.")

# ============================================================
# LOAD RETINANET (DETECTRON2)
# ============================================================

print("Loading RetinaNet...")

cfg = get_cfg()

cfg.merge_from_file(
    model_zoo.get_config_file(
        "COCO-Detection/retinanet_R_50_FPN_3x.yaml"
    )
)

cfg.MODEL.RETINANET.NUM_CLASSES = 7

cfg.MODEL.WEIGHTS = RETINA_WEIGHTS

cfg.MODEL.DEVICE = DEVICE

cfg.MODEL.RETINANET.SCORE_THRESH_TEST = RETINA_CONF

predictor = DefaultPredictor(cfg)

print("RetinaNet loaded successfully.")

# ============================================================
# LOAD IMAGE
# ============================================================

image_bgr = cv2.imread(IMAGE_PATH)

if image_bgr is None:
    raise ValueError(f"Could not read image: {IMAGE_PATH}")

image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

H, W = image_rgb.shape[:2]

# ============================================================
# IMAGE TRANSFORM
# ============================================================

transform = transforms.Compose([
    transforms.ToTensor()
])

tensor_image = transform(image_rgb).to(DEVICE)
```

## Detections

```python
# ============================================================
# YOLO DETECTION
# ============================================================

print("\nRunning YOLO detection...")

yolo_results = yolo_model.predict(
    source=image_rgb,
    conf=YOLO_CONF,
    verbose=False
)

yolo_boxes = []
yolo_scores = []
yolo_labels = []

for result in yolo_results:

    boxes = result.boxes.xyxy.cpu().numpy()
    scores = result.boxes.conf.cpu().numpy()
    labels = result.boxes.cls.cpu().numpy()

    for box, score, label in zip(boxes, scores, labels):

        x1, y1, x2, y2 = box

        box_area = (x2 - x1) * (y2 - y1)

        if box_area < MIN_BOX_AREA:
            continue

        yolo_boxes.append([
            x1 / W,
            y1 / H,
            x2 / W,
            y2 / H
        ])

        yolo_scores.append(float(score))
        yolo_labels.append(int(label))

# YOLO NMS

if len(yolo_boxes) > 0:

    yolo_boxes_tensor = torch.tensor([
        [
            box[0] * W,
            box[1] * H,
            box[2] * W,
            box[3] * H
        ]
        for box in yolo_boxes
    ])

    yolo_scores_tensor = torch.tensor(yolo_scores)

    keep = nms(
        yolo_boxes_tensor,
        yolo_scores_tensor,
        0.45
    )

    yolo_boxes = [yolo_boxes[i] for i in keep]
    yolo_scores = [yolo_scores[i] for i in keep]
    yolo_labels = [yolo_labels[i] for i in keep]

print(f"YOLO detections: {len(yolo_boxes)}")

# ============================================================
# FASTER R-CNN DETECTION
# ============================================================

print("\nRunning Faster R-CNN detection...")

with torch.no_grad():
    faster_outputs = faster_model([tensor_image])

faster_outputs = faster_outputs[0]

keep = nms(
    faster_outputs["boxes"],
    faster_outputs["scores"],
    0.25
)

faster_outputs["boxes"] = faster_outputs["boxes"][keep]
faster_outputs["scores"] = faster_outputs["scores"][keep]
faster_outputs["labels"] = faster_outputs["labels"][keep]

faster_boxes = []
faster_scores = []
faster_labels = []

for box, score, label in zip(
    faster_outputs["boxes"],
    faster_outputs["scores"],
    faster_outputs["labels"]
):

    score = score.item()

    if score < FASTER_CONF:
        continue

    x1, y1, x2, y2 = box.cpu().numpy()

    box_area = (x2 - x1) * (y2 - y1)

    if box_area < MIN_BOX_AREA:
        continue

    label = label.item() - 1

    if label < 0 or label >= len(CLASS_NAMES):
        continue

    faster_boxes.append([
        x1 / W,
        y1 / H,
        x2 / W,
        y2 / H
    ])

    faster_scores.append(score)
    faster_labels.append(label)

print(f"Faster detections: {len(faster_boxes)}")

# ============================================================
# RETINANET DETECTION
# ============================================================

print("\nRunning RetinaNet detection...")

outputs = predictor(image_bgr)

instances = outputs["instances"].to("cpu")

retina_boxes = []
retina_scores = []
retina_labels = []

boxes = instances.pred_boxes.tensor.numpy()
scores = instances.scores.numpy()
labels = instances.pred_classes.numpy()

for box, score, label in zip(
    boxes,
    scores,
    labels
):

    x1, y1, x2, y2 = box

    box_area = (x2 - x1) * (y2 - y1)

    if box_area < MIN_BOX_AREA:
        continue

    retina_boxes.append([
        x1 / W,
        y1 / H,
        x2 / W,
        y2 / H
    ])

    retina_scores.append(float(score))

    retina_labels.append(int(label))

print(f"RetinaNet detections: {len(retina_boxes)}")

# ============================================================
# HANDLE EMPTY
# ============================================================

if (
    len(yolo_boxes) == 0 and
    len(faster_boxes) == 0 and
    len(retina_boxes) == 0
):
    print("\nNo detections found.")
    exit()

# ============================================================
# WEIGHTED BOXES FUSION
# ============================================================

print("\nApplying Weighted Boxes Fusion...")

boxes_list = [
    yolo_boxes,
    faster_boxes,
    retina_boxes
]

scores_list = [
    yolo_scores,
    faster_scores,
    retina_scores
]

labels_list = [
    yolo_labels,
    faster_labels,
    retina_labels
]

fused_boxes, fused_scores, fused_labels = weighted_boxes_fusion(
    boxes_list,
    scores_list,
    labels_list,
    weights=[3, 1, 2],
    iou_thr=WBF_IOU,
    skip_box_thr=FINAL_CONF
)

# ============================================================
# FINAL FILTERING
# ============================================================

filtered_boxes = []
filtered_scores = []
filtered_labels = []

for box, score, label in zip(
    fused_boxes,
    fused_scores,
    fused_labels
):

    if score < FINAL_CONF:
        continue

    filtered_boxes.append(box)
    filtered_scores.append(score)
    filtered_labels.append(label)

fused_boxes = filtered_boxes
fused_scores = filtered_scores
fused_labels = filtered_labels

print(f"Final fused detections: {len(fused_boxes)}")
```

## Saving output

```python



# ============================================================
# DRAW RESULTS
# ============================================================

for box, score, label in zip(
    fused_boxes,
    fused_scores,
    fused_labels
):

    x1 = int(box[0] * W)
    y1 = int(box[1] * H)
    x2 = int(box[2] * W)
    y2 = int(box[3] * H)

    label = int(label)

    if label not in CLASS_NAMES:
        continue

    class_name = CLASS_NAMES[label]

    text = f"{class_name} {score:.2f}"

    cv2.rectangle(
        image_bgr,
        (x1, y1),
        (x2, y2),
        (0, 255, 0),
        2
    )

    cv2.putText(
        image_bgr,
        text,
        (x1, y1 - 10),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.7,
        (0, 255, 0),
        2
    )

# ============================================================
# SAVE OUTPUT
# ============================================================

OUTPUT_PATH = "fusion_output.jpg"

cv2.imwrite(OUTPUT_PATH, image_bgr)

print(f"\nSaved output to: {OUTPUT_PATH}")

```

## Cropped_Images

```python
# ============================================================
# CROP DETECTIONS FOR CLASSIFIERS
# ============================================================

print("\nCropping fused detections for classifiers...")

cropped_images = []

for i, box in enumerate(fused_boxes):

    # Convert normalized coords back to image coords
    x1 = int(box[0] * W)
    y1 = int(box[1] * H)
    x2 = int(box[2] * W)
    y2 = int(box[3] * H)

    # Safety clipping
    x1 = max(0, x1)
    y1 = max(0, y1)
    x2 = min(W, x2)
    y2 = min(H, y2)

    # Crop object
    crop = image_rgb[y1:y2, x1:x2]

    # Skip invalid crops
    if crop.size == 0:
        continue

    cropped_images.append({
        "crop": crop,
        "box": [x1, y1, x2, y2],
        "score": fused_scores[i],
        "label": fused_labels[i]
    })

print(f"Total cropped objects: {len(cropped_images)}")
```

## Classifiers

```python
# ============================================================
# CLASSIFICATION FUSION FUNCTION
# ============================================================

def classify_crop_ensemble(crop_rgb):

    # RGB -> PIL
    image_pil = Image.fromarray(crop_rgb)

    # Transform
    tensor = transform(image_pil).unsqueeze(0).to(device)

    with torch.no_grad():

        # ====================================================
        # RESNET50
        # ====================================================

        resnet_output = resnet_model(tensor)

        resnet_probs = torch.softmax(
            resnet_output,
            dim=1
        )

        resnet_class = torch.argmax(
            resnet_probs,
            dim=1
        ).item()

        resnet_conf = resnet_probs[0][
            resnet_class
        ].item()

        # ====================================================
        # CONVNEXT
        # ====================================================

        convnext_output = convnext_model(tensor)

        convnext_probs = torch.softmax(
            convnext_output,
            dim=1
        )

        convnext_class = torch.argmax(
            convnext_probs,
            dim=1
        ).item()

        convnext_conf = convnext_probs[0][
            convnext_class
        ].item()

        # ====================================================
        # EFFICIENTNET
        # ====================================================

        efficient_output = efficient_model(tensor)

        efficient_probs = torch.softmax(
            efficient_output,
            dim=1
        )

        efficient_class = torch.argmax(
            efficient_probs,
            dim=1
        ).item()

        efficient_conf = efficient_probs[0][
            efficient_class
        ].item()

        # ====================================================
        # WEIGHTED PROBABILITY FUSION
        # ====================================================

        fused_probs = (

            (RESNET_WEIGHT * resnet_probs) +

            (CONVNEXT_WEIGHT * convnext_probs) +

            (EFFICIENT_WEIGHT * efficient_probs)

        ) / (

            RESNET_WEIGHT +

            CONVNEXT_WEIGHT +

            EFFICIENT_WEIGHT
        )

        # ====================================================
        # FINAL FUSED PREDICTION
        # ====================================================

        final_class = torch.argmax(
            fused_probs,
            dim=1
        ).item()

        final_conf = fused_probs[0][
            final_class
        ].item()

    return {

        # FINAL RESULT
        "class_id": final_class,

        "class_name": CLASS_NAMES[final_class],

        "confidence": final_conf,

        # RESNET
        "resnet_class": CLASS_NAMES[resnet_class],

        "resnet_conf": resnet_conf,

        # CONVNEXT
        "convnext_class": CLASS_NAMES[convnext_class],

        "convnext_conf": convnext_conf,

        # EFFICIENTNET
        "efficient_class": CLASS_NAMES[efficient_class],

        "efficient_conf": efficient_conf
    }

# ============================================================
# RUN CLASSIFIER ENSEMBLE
# ============================================================

print("\nRunning classifier ensemble...")

final_results = []

for i, item in enumerate(cropped_images):

    crop = item["crop"]

    result = classify_crop_ensemble(crop)

    final_results.append({

        "box": item["box"],

        "detection_score": item["score"],

        "detection_label": item["label"],

        "classification": result
    })

    print(f"\n================ OBJECT {i+1} ================")

    print(
        f"ConvNext     >> "
        f"{result['convnext_conf']:.4f} >> "
        f"{result['convnext_class']}"
    )

    print(
        f"ResNet50     >> "
        f"{result['resnet_conf']:.4f} >> "
        f"{result['resnet_class']}"
    )

    print(
        f"EfficientNet >> "
        f"{result['efficient_conf']:.4f} >> "
        f"{result['efficient_class']}"
    )

    print("\n------------- FINAL FUSION -------------")

    print(
        f"Final Class  >> "
        f"{result['confidence']:.4f} >> "
        f"{result['class_name']}"
    )
```
