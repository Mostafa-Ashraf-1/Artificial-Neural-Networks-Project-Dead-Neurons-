```python
import json

def fix_coco(input_path, output_path):
    with open(input_path, 'r') as f:
        data = json.load(f)

    fixed_annotations = []
    removed = 0

    for ann in data["annotations"]:
        bbox = ann["bbox"]

        # Convert to float
        try:
            x, y, w, h = map(float, bbox)
        except:
            removed += 1
            continue

        # Remove invalid boxes
        if w <= 0 or h <= 0:
            removed += 1
            continue

        ann["bbox"] = [x, y, w, h]
        fixed_annotations.append(ann)

    data["annotations"] = fixed_annotations

    with open(output_path, 'w') as f:
        json.dump(data, f)

    print(f"✅ Fixed file saved")
    print(f"❌ Removed {removed} invalid annotations")

annotations_path = "Datasets/Ragab.coco/train/_annotations.coco.json"
output = "Datasets/Ragab.coco/train/_annotations_fixed.coco.json"

fix_coco(annotations_path, output)
```
