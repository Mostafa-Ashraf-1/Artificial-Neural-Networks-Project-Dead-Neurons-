```python
from ultralytics import YOLO
import os

# 1. Load a pretrained YOLOv8 model (small version is good for testing)
model = YOLO('runs/detect/garbage_detection_test/weights/last.pt')

# 2. Train the model using your data.yaml
# We set epochs to a low number (like 5 or 10) just to check if the pipeline works
results = model.train(
    resume=True,  # resume from last checkpoint
    data='data.yaml',
    epochs=100,
    imgsz=640,
    batch=-1, #automatically adjust batch size for approximately 60% CUDA memory utilization.
    name='garbage_detection_test',
    patience=0,         # disable early stopping
    save_period=10,     # checkpoint every 10 epochs, so you don't lose progress
    workers=4,          # parallel data loading; tune to your CPU core count
    device=0,           # explicitly use GPU 0; use 'cpu' if no GPU
    exist_ok=True,      # won't crash if the run name already exists
    cache=True,
    cls_pw=0.5,
)

# 3. Validate the model's progress on the validation set
best_model = YOLO('runs/detect/garbage_detection_test/weights/best.pt')
metrics = best_model.val()
print(f"mAP50-95: {metrics.box.map:.4f}")
print(f"mAP50:    {metrics.box.map50:.4f}")

# 4. Run inference on a sample test image to visualize results
# Replace 'test_image.jpg' with an actual path to one of your images
# results_inf = model.predict('dataset1/test/images/000034_jpg.rf.3G4te800F18CbzoiJ2qK.jpg', save=True)
```
