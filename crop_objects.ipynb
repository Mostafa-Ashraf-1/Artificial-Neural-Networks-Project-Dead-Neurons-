```python
import os
import cv2
import json

# ===== PATHS =====
images_dir = "Dataset/test"
annotation_file = "Dataset/test/_annotations.coco.json"

output_dir = "classifier_dataset/test"

# ===== LOAD COCO JSON =====
with open(annotation_file, "r") as f:
    coco = json.load(f)

# ===== CREATE CATEGORY MAPPING =====
# Example:
# {1: "plastic", 2: "metal", ...}

categories = {
    cat["id"]: cat["name"]
    for cat in coco["categories"]
}

# ===== CREATE OUTPUT FOLDERS =====
for class_name in categories.values():
    os.makedirs(
        os.path.join(output_dir, class_name),
        exist_ok=True
    )

# ===== CREATE IMAGE ID LOOKUP =====
# Example:
# {12: "abc.jpg"}

image_lookup = {
    img["id"]: img["file_name"]
    for img in coco["images"]
}

# ===== PROCESS ANNOTATIONS =====
for idx, ann in enumerate(coco["annotations"]):

    image_id = ann["image_id"]
    category_id = ann["category_id"]

    # Get image filename
    image_name = image_lookup[image_id]

    image_path = os.path.join(images_dir, image_name)

    # Read image
    image = cv2.imread(image_path)

    if image is None:
        continue

    # COCO bbox format:
    # [x, y, width, height]

    x, y, w, h = ann["bbox"]

    x1 = int(x)
    y1 = int(y)
    x2 = int(x + w)
    y2 = int(y + h)

    # Clamp coordinates
    img_h, img_w = image.shape[:2]

    x1 = max(0, x1)
    y1 = max(0, y1)
    x2 = min(img_w, x2)
    y2 = min(img_h, y2)

    # Crop object
    crop = image[y1:y2, x1:x2]

    # Skip invalid crops
    if crop.size == 0:
        continue

    # Class name
    class_name = categories[category_id]

    # Save filename
    base_name = os.path.splitext(image_name)[0]

    save_name = f"{base_name}_{idx}.jpg"

    save_path = os.path.join(
        output_dir,
        class_name,
        save_name
    )

    # Save crop
    cv2.imwrite(save_path, crop)

print("Done!")
```
