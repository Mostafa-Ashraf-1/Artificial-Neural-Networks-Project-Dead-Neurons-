```python
import os
import csv

# Root dataset folder
dataset_path = r"classifier_dataset/train"

# Output CSV file
output_csv = "classifier_dataset/train.csv"

# Folder -> Target mapping
class_map = {
    "Cardboard": 1,
    "Clothes": 2,
    "Garbage_bag": 3,
    "Glass": 4,
    "Metal": 5,
    "Organic": 6,
    "Plastics": 7
}

# Supported image extensions
image_extensions = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

rows = []

# Loop through each class folder
for folder_name, label in class_map.items():

    folder_path = os.path.join(dataset_path, folder_name)

    if not os.path.exists(folder_path):
        print(f"Folder not found: {folder_path}")
        continue

    # Walk through all images inside the folder
    for root, dirs, files in os.walk(folder_path):
        for file in files:

            if file.lower().endswith(image_extensions):

                image_path = os.path.join(root, file)

                # Save row: image path + label
                rows.append([image_path, label])

# Write CSV
with open(output_csv, mode="w", newline="", encoding="utf-8") as f:

    writer = csv.writer(f)

    # Header
    writer.writerow(["image_path", "target"])

    # Data
    writer.writerows(rows)

print(f"CSV file saved as: {output_csv}")
print(f"Total images: {len(rows)}")
```
